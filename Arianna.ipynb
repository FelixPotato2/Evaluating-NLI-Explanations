```python
import json

# ---- 0) Load your example inputs ----
# If you already have `responses` in memory, skip loading it.
# Otherwise:
with open("final_LLM_auto_responses.json", "r", encoding="utf-8") as f:
    responses = json.load(f)

with open("annotators_answers.json", "r", encoding="utf-8") as f:
    answers_all = json.load(f)

pairID = '3089862485.jpg#0r1e'

# LLM output for this pair (dict with keys like "answer", "explanation")
llm_one = {pairID: responses["q0727"]}   # <-- you said responses['q0011'] corresponds to this pair after mapping
# If instead you already have the full-id-keyed output, use:
# llm_one = {pairID: responses[pairID]}

# Gold answers for this pair (list of answer groups)
gold_one = {pairID: answers_all[pairID]}

print("LLM raw explanations:")
print(llm_one[pairID].get("explanation", []))
print("\nGold answer groups:")
print(gold_one[pairID])
```

```python
def subset_with_extra_limit(pred_tokens, gold_tokens, max_extra=2):
    pred_set = set(pred_tokens)
    gold_set = set(gold_tokens)

    if not gold_set.issubset(pred_set):
        return False, None

    extras = len(pred_set - gold_set)
    return extras <= max_extra, extras

def _extract_typeof_relations(explanations):
    """Return only explanations that contain 'is a type of'."""
    if not explanations:
        return []
    out = []
    for ex in explanations:
        if isinstance(ex, str) and "is a type of" in ex.lower():
            out.append(ex)
    return out

def calculate_scores(result, key):
    TP = sum(v[key] for v in result.values() if isinstance(v, dict))
    FP = sum(v["total_LLM_answers"] - v[key] for v in result.values() if isinstance(v, dict))
    FN = sum(v["total_answers"] - v[key] for v in result.values() if isinstance(v, dict))

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
    recall    = TP / (TP + FN) if (TP + FN) > 0 else 0.0
    f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0.0
    return {"precision": precision, "recall": recall, "f1": f1}

def calculate_counts(result, key):
    TP = sum(v[key] for v in result.values() if isinstance(v, dict))
    FP = sum(v["total_LLM_answers"] - v[key] for v in result.values() if isinstance(v, dict))
    FN = sum(v["total_answers"] - v[key] for v in result.values() if isinstance(v, dict))
    return {"TP": TP, "FP": FP, "FN": FN}

```

```python
def check_LLM_answer(answers, LLM_output, max_extra_total=2):
    """
    Adds a new metric (len_ok) where an LLM relation is correct if:
      - gold-left âŠ† llm-left AND gold-right âŠ† llm-right
      - (extra_left + extra_right) <= max_extra_total

    IMPORTANT (fixed):
      exact âŠ† len_ok âŠ† partial  (counts are inclusive now)
    """
    pairs_with_no_typeof = 0
    pairs_with_no_output = 0

    result = {}
    for pairID in answers:
        print("\n" + "="*80)
        print("PAIR:", pairID)

        if pairID not in LLM_output:
            print("-> No LLM output for this pairID")
            result[pairID] = {
                "exact": 0, "partial": 0, "len_ok": 0,
                "combined_correct": 0, "combined_len_ok": 0,
                "total_answers": len(answers[pairID]),
                "total_LLM_answers": 0,
            }
            pairs_with_no_output += 1
            continue

        exact_count = 0
        partial_count = 0
        len_ok_count = 0

        raw_explanations = LLM_output[pairID].get("explanation", [])
        llm_explanations = _extract_typeof_relations(raw_explanations)

        print("Raw explanations:", raw_explanations)
        print("Filtered (type-of) explanations:", llm_explanations)

        if len(llm_explanations) == 0:
            pairs_with_no_typeof += 1

        for idx, LLM_answer in enumerate(llm_explanations, start=1):
            print("\n--- LLM relation", idx, "---")
            print("LLM_answer:", LLM_answer)

            LLM_answer = (LLM_answer or "").lower()
            splitted_ans = LLM_answer.split("is a type of")
            if len(splitted_ans) < 2:
                print("-> skipped (no split)")
                continue

            articles = {"a", "an", "the"}
            spl_str_ans = [
                [tok.strip(" ,.") for tok in seg.lower().split()
                 if tok.strip(" ,.").lower() not in articles]
                for seg in splitted_ans[:2]
            ]

            llm_left = spl_str_ans[0]
            llm_right = spl_str_ans[1]
            print("Tokenized llm_left :", llm_left)
            print("Tokenized llm_right:", llm_right)

            found_exact = False
            found_len_ok = False
            found_partial = False

            for g_i, answer_group_dict in enumerate(answers[pairID], start=1):
                print(f"\n  Gold group {g_i}: {answer_group_dict}")

                left_exact = right_exact = False
                left_partial = right_partial = False

                left_extras_candidates = []
                right_extras_candidates = []

                # Left side
                for left_answer in answer_group_dict.get("left", []):
                    if left_answer == llm_left:
                        left_exact = True
                    if bool(set(left_answer) & set(llm_left)):
                        left_partial = True

                    okL, extraL = subset_with_extra_limit(llm_left, left_answer, max_extra=max_extra_total)
                    if okL:
                        left_extras_candidates.append(extraL)

                # Right side
                for right_answer in answer_group_dict.get("right", []):
                    if right_answer == llm_right:
                        right_exact = True
                    if bool(set(right_answer) & set(llm_right)):
                        right_partial = True

                    okR, extraR = subset_with_extra_limit(llm_right, right_answer, max_extra=max_extra_total)
                    if okR:
                        right_extras_candidates.append(extraR)

                print("    left_exact/right_exact:", left_exact, right_exact)
                print("    left_partial/right_partial:", left_partial, right_partial)
                print("    left_extras_candidates:", left_extras_candidates)
                print("    right_extras_candidates:", right_extras_candidates)

                if left_exact and right_exact:
                    print("    -> FOUND EXACT in this group")
                    found_exact = True
                    found_len_ok = True    # exact implies len_ok
                    found_partial = True   # exact implies partial
                    break

                if left_extras_candidates and right_extras_candidates:
                    best_total = min(left_extras_candidates) + min(right_extras_candidates)
                    print("    best_total_extras:", best_total, "(limit:", max_extra_total, ")")
                    if best_total <= max_extra_total:
                        print("    -> FOUND LEN_OK in this group")
                        found_len_ok = True
                        found_partial = True  # len_ok implies partial

                if left_partial and right_partial:
                    print("    -> FOUND PARTIAL in this group")
                    found_partial = True

            # INCLUSIVE COUNTING (fixed)
            if found_exact:
                exact_count += 1
                len_ok_count += 1
                partial_count += 1
                print("=> counts: exact +1, len_ok +1, partial +1")
            elif found_len_ok:
                len_ok_count += 1
                partial_count += 1
                print("=> counts: len_ok +1, partial +1")
            elif found_partial:
                partial_count += 1
                print("=> counts: partial +1")
            else:
                print("=> counts: no match")

        # FIXED combined_*: do NOT sum nested counts
        result[pairID] = {
            "exact": exact_count,
            "partial": partial_count,
            "len_ok": len_ok_count,
            "combined_correct": partial_count,   # loose = partial-or-better
            "combined_len_ok": len_ok_count,     # len_ok-or-better
            "total_answers": len(answers[pairID]),
            "total_LLM_answers": len(llm_explanations),
        }

        print("\nPAIR RESULT:", result[pairID])

    scores_strict = calculate_scores(result, "exact")
    scores_loose = calculate_scores(result, "combined_correct")
    len_scores   = calculate_scores(result, "combined_len_ok")
    len_metrics  = calculate_counts(result, "combined_len_ok")

    print("\n" + "="*80)
    print("Pairs with no type of:", pairs_with_no_typeof)
    print("Pairs with empty/no output:", pairs_with_no_output)
    print("STRICT:", scores_strict)
    print("LOOSE :", scores_loose)
    print("LEN_OK:", len_scores, "counts:", len_metrics)

    return result, scores_strict, scores_loose, len_scores, len_metrics
```

```python
result, scores_strict, scores_loose, len_scores, len_metrics = check_LLM_answer(gold_one, llm_one, max_extra_total=2)
```
