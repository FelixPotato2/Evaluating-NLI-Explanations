```python
def print_example(df, ID = None, rownum = None, ):
    
    if ID is not None: 
        row = df.loc[df["pairID"] == ID]
        if row.empty:
            print("ID not found.")
            return
        row = row.iloc[0]  
    if rownum is not None:
        row = df.iloc[rownum]
    for col in df.columns:
        #if "Highlighted" in col: 
            #continue
        #else: 
        print(f"{col}:")
        print(row[col])
        print()
```

```python
df_glob = pd.read_csv("entailment_probs_or.csv")
#df = pd.read_csv("processed_esnli_EA.csv")
#df = pd.read_csv("esnli_dev.csv")
print(f"amount of rows:{df_glob.shape[0]}")        
for i in range(0,10):
    print(f"----------------EXAMPLE {i} ----------------\n")
    print_example(df_glob, rownum=i)
```
